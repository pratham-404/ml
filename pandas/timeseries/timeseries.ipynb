```python
df = pd.read_csv("./data/aapl_no_dates.csv")
# df

# we can use date range incase our dataset does not contain date as a column
range = pd.date_range(start="6/1/2017", end="6/30/2017", freq='B')
# range

df.set_index(range, inplace=True)
df
```

```python
# fillup holes in dataset using previous values usgin asfreq() method
df.asfreq('D', method='pad') # observe new index on weekends are created and set ot values of previous date's available data
```

```python
# we use shift mostly to calculate percentage change in prices
df.shift(1).head()
```

```python
df.shift(-1).tail()
```

```python
df['Prev. Close'] = df['Close'].shift(1)
df.head()
```

```python
df["1 day change"] = df['Close'] - df["Prev. Close"]
df.head()
```

```python
df["5 day % change"] = (df['Close'] - df["Close"].shift(5))*100/(df["Close"].shift(5)) # smilarly we can also shift he index using tshift() in pandas
df.head()
```
