### Data Visualization using matplotlib

#### Line Charts

```python
# pip install matplotlib

import matplotlib.pyplot as plt
# %matplotlib inline
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y, color="green", linewidth=7, linestyle='dotted')
```

```python
plt.plot(x, y, marker='D', color='blue', alpha=0.25, markersize=10) # alpha => transparency
```

```python
plt.plot(x,y,'r+--') # format strings (r+-- => color => red, marker => +, linestyle => --(dotted))
```

#### Bar Charts

```python
import numpy as np

# Bar Chart
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

# Axes
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Stocks")

# y-position
xpos = np.arange(len(company))
plt.xticks(xpos, company) # X axis is fixed with company names

# multiple values can be shown in a single plot

# plt.bar(ypos,revenue, label='company')
# plt.bar(ypos,profit, label='profit')

plt.bar(xpos-0.2, revenue, width=0.4, label='company')
plt.bar(xpos+0.2, profit, width=0.4, label='profit')

plt.legend()
```

```python
plt.xlabel("Company")
plt.ylabel("Revenue")
plt.title("US Stocks")

plt.yticks(xpos, company)

plt.barh(xpos-0.2, revenue, height=0.4, label='company')
plt.barh(xpos+0.2, profit, height=0.4, label='profit')

plt.legend()
```

#### Histograms

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

# plt.hist(blood_sugar) # by default => bins(10)
# plt.hist(blood_sugar, bins=5)
# plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95)
plt.hist(blood_sugar, bins=[80,100,125,150], rwidth=0.95, histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.xlabel('Blood sugar level (mg)')
plt.ylabel('No. of persons (frequency)')
plt.title('Blood Sugar Statistics')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], color=['r','b'], label=['men', 'women'])
# plt.hist([blood_sugar_men, blood_sugar_women], bins=[80, 100, 125, 150], color=['r','b'], label=['men', 'women'], orientation='horizontal')
plt.legend()
```

#### Pie Charts

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent", "Food", "Phone/Internet Bill", "Car", "Other Utilities"]

plt.title("Monthly expense chart")
plt.pie(exp_vals, labels=exp_labels, radius=1.1, autopct="%0.2f%%", explode=[0,0.1,0,0.1,0], startangle=60)

plt.savefig('./figures/pie.png', bbox_inches='tight', pad_inches=1, transparent=True)
plt.savefig('./figures/pie.pdf', bbox_inches='tight', pad_inches=1, transparent=True)
```
