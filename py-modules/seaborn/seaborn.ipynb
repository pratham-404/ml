```python
import seaborn as sns
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
df = pd.read_csv('./data/tips.csv')
df.head()
```

```python
# Barplot
sns.barplot(x='day', y='total_bill', data=df)
```

```python
# Hue Attribute => legend + different column in each entry in x-axis
sns.barplot(x='day', y='total_bill', hue='sex', data=df)
```

```python
# color pallete(deep, light, winter_r, spring, etc)
sns.palplot(sns.color_palette('deep', 10))
```

```python
# Custom pallete in seaborn
custom = ['red', 'yellow', 'green', 'blue']
sns.set_palette(custom)
sns.palplot(sns.color_palette())
```

```python
# Horizontal Bar plot
sns.barplot(x='total_bill', y='day', hue='sex', data=df)
```

```python
# reorganizing bars 
sns.barplot(x='day', y='total_bill', order=['Sun', 'Fri', 'Thur', 'Sat'], data=df)
```

```python
# Saving plot as an image
sns_plot = sns.barplot(x='day', y='total_bill', hue='sex', data=df)
fig = sns_plot.get_figure()
fig.savefig(fname='./files/barplot.png')
```
